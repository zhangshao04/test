```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [0,1,2,3]
y = [0,1,4,9]
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-5,5,100)
plt.plot(x,x**2,'r--')
plt.plot(x,x**3,'g^')
plt.show()
```

```python
plt.figure(figsize=(10,6))#画布大小
x = np.linspace(-5,5,100)
plt.plot(x,x**2,'r--',label = 'x**2')
plt.plot(x,x**3,'g^',label = 'x**3')
plt.title(f'Plot of Various Polynomials from {x[0]} to {x[-1]}')
plt.xlabel('x Axis',fontsize = 18)#fontsize用来改变标签大小
plt.ylabel('y Axis',fontsize = 18)
plt.legend(loc = 0)
plt.xlim(-4,4)
plt.ylim(-10,10)
plt.grid()#打开图像网格
plt.show()
```

```python
x = np.arange(11)
y = x**2

plt.figure(figsize = (14, 8))

plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.title('Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.subplot(2, 3, 2)
plt.scatter(x,y)
plt.title('Scatter')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.subplot(2, 3, 3)
plt.bar(x,y)
plt.title('Bar')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.subplot(2, 3, 4)
plt.loglog(x,y)
plt.title('Loglog')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(which='both')

plt.subplot(2, 3, 5)
plt.semilogx(x,y)
plt.title('Semilogx')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(which='both')

plt.subplot(2, 3, 6)
plt.semilogy(x,y)
plt.title('Semilogy')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.tight_layout()#让子图彼此互不重叠

plt.show()
```
